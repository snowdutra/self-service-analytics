Lendo arquivo **csv** com **Pandas**

```python
df = pd.read_csv("entrada.csv")
df
```

Obter colunas específicas do **Dataframe** importado

```python
df[["nome", "sistema_origem"]]
```

Obter valores únicos, total de itens e somatória (veja que estará errado pois nem todos valores são numéricos!)

```python
print(df["sistema_origem"].unique())
print(df["id_cliente"].count())
print(df["quantidade"].sum())
```

Removendo espaços extras com **strip**

```python
df["id_cliente"] = df["id_cliente"].str.strip()
df["nome"] = df["nome"].str.strip()
```

Padronizando maiúsculo / minúsculo

```python
df["nome"] = df["nome"].str.title()
df["sistema_origem"] = df["sistema_origem"].str.upper()
df["produto"] = df["produto"].str.upper()
```

Remover dados faltantes, por exemplo **id_cliente**

```python
df = df.dropna(subset="id_cliente")
```

Trabalhando números tratados como texto, por exemplo na quantidade

```python
df["quantidade"] = pd.to_numeric(df["quantidade"], errors="coerce")
df = df.dropna(subset="quantidade")
```

Criar uma função para trabalhar as datas em diversos formatos

```python
def parse_data(valor):
  formatos = ["%Y-%m-%d", "%d/%m/%Y", "%Y/%m/%d"]
  for f in formatos:
    try:
      return pd.to_datetime(valor, format=f)
    except:
      pass
  return pd.NaT
```

Aplicar a função **parse_data** e eliminar as linhas fora do padrão

```python
df['data'] = df['data'].apply(parse_data)
df = df.dropna(subset="data")
```

Removendo linhas duplicadas (por exemplo linhas 2 e 3)

```python
df = df.drop_duplicates()
df
```

Criando colunas derivadas, por exemplo, total como cálculo de quantidade * preco

```python
df['total_calculado'] = df['quantidade'] * df['preco']
df
```
