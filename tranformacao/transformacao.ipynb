```python
import pandas as pd

df = pd.read_csv('transformacao.csv', sep=';')

df.head()
```

```python
total_registros = len(df)
print(f'Total de registros: {total_registros}')
```

```python
campos_nulos = df["placa"].isnull().sum()
completude = campos_nulos / total_registros
print(f"Placa nula (%) = {100 * completude}%")
```

```python
estados_validos = [
"AC","AL","AP","AM","BA","CE","DF","ES","GO",
"MA","MT","MS","MG","PA","PB","PR","PE","PI",
"RJ","RN","RS","RO","RR","SC","SP","SE","TO"
]
```

```python
estados_invalidos = (~df["estado"].isin(estados_validos)).sum()
validity = estados_invalidos / total_registros
print(f"Estados inválidos (%) = {100 * validity}%")
```

```python
df["valor_calculado"] = df["valor_venda"] + df["taxa_corretagem"]
consistency = (df["total_venda"] != df["valor_calculado"]).sum() / total_registros
print(f"Valores inconsistentes (%) = {round(consistency*100,2)}%")
```

```python
duplicados = df.duplicated()
```

```python
uniqueness = (total_registros - df["registro_id"].nunique()) / total_registros
print(uniqueness)
print(f"Id duplicados (%) = {round(uniqueness*100,2)}%")
```

```python
df["data_venda"] = pd.to_datetime(df["data_venda"])
df["idade_dias"] = (pd.Timestamp.now() - df["data_venda"]).dt.days
print(df["idade_dias"].describe())
```

```python
modelos_validos = []

integridade = df["modelo_id"].isin(modelos_validos)
taxa_integridade = integridade.mean()
print("Taxa de integridade:", round(taxa_integridade * 100, 2), "%")

df_integridade_ruim = df.loc[~integridade, ["registro_id", "marca_id", "modelo_id"]]
print(df_integridade_ruim)
```
